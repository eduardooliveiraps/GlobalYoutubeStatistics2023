## Data Exploration

```python
# Load the dataset from a CSV file into a Pandas DataFrame
df = pd.read_csv("../data/supermarket_sales.csv")
```

## Data Preperation

```python
# Convert the 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'])
```

```python
# Rename the 'cogs' column to 'COGS'
df.rename(columns={'cogs':'COGS'})
```

## Data Analysis

### Question 1: What is the total revenue for the entire dataset?

```python
print(f"Total revenue: {round(df['gross income'].sum(),2)} USD")
```

### Question 2: How does revenue vary across different cities?

```python
# Calculate the sum of 'gross income' for each city
df.groupby('City')['gross income'].sum().round(2)
```

```python
revenue_per_city = df.groupby('City')['gross income'].sum().round(2)
```

```python
# Create a bar plot to visualize revenue per city
fig = revenue_per_city.plot.bar()
fig.set_title('Revenue by City')
fig.set_xlabel('City')
fig.set_ylabel('Revenue (USD)')
```

### Question 3: What is the distribution of customer types (Member vs. Normal)?

```python
df['Customer type'].value_counts()
```

```python
# Create a pie chart to visualize the distribution of customer types
customer_type = df['Customer type'].value_counts()
fig = customer_type.plot.pie(autopct='%1.1f%%', labels = ['Member', 'Normal'])
fig.set_ylabel('')
fig.set_title('Customer type distribution')
```

### Question 4: How does gender affect purchase behavior and total sales?

```python
# Average amount spent per sale by gender
df.groupby('Gender')['Total'].mean().round(2)
```

```python
# Total amount spent on sales by gender
df.groupby('Gender')['Total'].sum().round(2)
```

```python
# Pivot_table of Total Sales by Product Category and Gender
total_amount_pivot_table = df.pivot_table(index = 'Product line', columns = 'Gender', values = 'Total', aggfunc = 'sum').round(2)
```

```python
# Create a bar plot to visualize total sales by product category and gender
fig = total_amount_pivot_table.plot.bar()
fig.set_xlabel('Product Category')
fig.set_ylabel('Total Sales')
fig.set_title('Total Sales by Product Category and Gender')
fig.legend()
```

### Question 5: Which product line has the highest sales and gross income?

```python
# Create a pivot table to calculate the sum of 'Total' and 'gross income' for each product line
sales_income_product = df.pivot_table(index = 'Product line', values = ['Total','gross income'], aggfunc = 'sum').round(2)
```

```python
# Create a stacked bar plot to visualize total sales and gross income by product category
fig = sales_income_product.plot(kind = 'bar', stacked = True)
fig.legend(['Total Sales','Gross Income'], bbox_to_anchor=(1.3, 1), loc = 'upper right', borderaxespad=0)
fig.set_title('Sales and Gross Income by Product Category')
fig.set_xlabel('Product Category')
fig.set_ylabel('Sales and Gross Income')
```

```python
# Group the data by 'Product line' and calculate the sum of 'gross income' for each product category
income_by_product = df.groupby('Product line')['gross income'].sum().round(2).sort_values()

fig = income_by_product.plot(kind='barh', title='Gross Income by Product Category', grid=True)

# Annotate bars with values
for index, value in enumerate(income_by_product):
    fig.text(value + 10, index, f'{value:,}', va='center')

fig.set_ylabel('Product Category')
```

### Question 6: What is the distribution of sales throughout the week?

```python
# Insert a new column 'Weekday'
weekday_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['Date'] = pd.to_datetime(df['Date'])
df.insert(loc = 11, column = 'Weekday', value = df['Date'].apply(lambda x: weekday_names[x.weekday()]))
df.head()
```

```python
# Group the data by 'Weekday' and calculate the sum of 'Total' sales for each weekday
week_sales = df.groupby('Weekday')['Total'].sum().round(3)
week_sales = week_sales.reindex(weekday_names)
```

```python
# Create a line plot to visualize total sales by weekday
fig = week_sales.plot(kind = 'line', marker = 'o')
fig.set_ylabel('Total Sales')
fig.set_title('Total Sales by Weekday')
```

### Question 7: Does the time of day impact purchase behavior?

```python
# Insert a new column 'Purchase Hour'
df.insert(loc = 13, column = 'Purchase Hour', value = df['Time'].apply(lambda x: x.hour))
df.head()
```

```python
mean_sales_by_hour = df.groupby('Purchase Hour')['Total'].mean().round(3)
```

```python
fig = mean_sales_by_hour.plot(kind = 'line', marker= 'o', xticks = range(7,23), title = 'Average Sales by Hour of the Day', ylabel = 'Average Sales')
```

### Question 8: Which payment method is most commonly used?

```python
# Count the occurrences of each payment method in the 'Payment' column
payment_method = df['Payment'].value_counts()
```

```python
fig = payment_method.plot(kind = 'bar', grid = True, title = 'Most Used Payment Method', ylabel = 'Total value')
```

### Question 9: Is there a relationship between unit price and Sold Quantity?

```python
# Create a scatter plot to visualize the relationship between unit price and sold quantity
plt.scatter(df['Unit price'], df['Quantity'], marker = 'o')
plt.title('Relationship between Unit Price and Sold Quantity')
plt.xlabel('Unit Price')
plt.ylabel('Sold Quantity')
plt.grid(True)
```

```python
# Correlation Analysis
df.corr()
```

Based on the correlation coefficient of 0.010778 between Unit price and Quantity, it seems that there is almost no meaningful linear relationship between unit price and quantity of items sold. This suggests that changes in unit price do not strongly impact the quantity of items customers purchase 

### Question 10: What is the average quantity of items purchased in each transaction?

```python
mean_value = df['Quantity'].mean()
print(f'In each transaction the average quantity of items purchased is: {round(mean_value,0)}')
```
